```python

data['stress_bucket'] = pd.cut(data['stress_level'], bins=[0, 4, 7, 10], labels=['Low Stress (1-4)', 'Medium Stress (4-7)', 'High Stress (7-10)'])
data['self_esteem_bucket'] = pd.cut(data['self_esteem'], bins=[0, 4, 7, 10], labels=['Low Self-Esteem (1-4)', 'Medium Self-Esteem (4-7)', 'High Self-Esteem (7-10)'])

 
stress_analysis = data.groupby('stress_bucket' , observed=False)[['monthly_purchase_amount', 'total_orders']].mean()
print("---analysis of stress level buckets---")
stress_analysis
```

```python
esteem_analysis = data.groupby('self_esteem_bucket' , observed=False)[['monthly_purchase_amount', 'average_order_value']].mean()
print("---analysis of self esteem level buckets---")
esteem_analysis
```
